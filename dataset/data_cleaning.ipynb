## Preproccessing

### Uporządkowanie kolumn

```python
credits.rename(columns={'id': 'movieId'}, inplace=True)
keywords.rename(columns={'id': 'movieId'}, inplace=True)
```

```python
credits['movieId'] = pd.to_numeric(credits['movieId'], errors='coerce').astype('Int64')
keywords['movieId'] = pd.to_numeric(keywords['movieId'], errors='coerce').astype('Int64')
links['movieId'] = pd.to_numeric(links['movieId'], errors='coerce').astype('Int64')
links['tmdbId'] = pd.to_numeric(links['tmdbId'], errors='coerce').astype('Int64')
# ratings['movieId']= pd.to_numeric(ratings['movieId'], errors='coerce').astype('Int64')
movies['id'] = pd.to_numeric(movies['id'], errors='coerce').astype('Int64')
```

```python
movies.columns
```

```python
movies.drop('belongs_to_collection', inplace=True, axis=1)
movies.drop('homepage', inplace=True, axis=1)
movies.drop('original_title', inplace=True, axis=1)
movies.drop('revenue', inplace=True, axis=1)
movies.drop('status', inplace=True, axis=1)
movies.drop('video', inplace=True, axis=1)
movies.drop('spoken_languages', inplace=True, axis=1)
movies.drop('imdb_id', inplace=True, axis=1)
movies.drop('production_companies', inplace=True, axis=1)
movies.drop('production_countries', inplace=True, axis=1)
movies.drop('tagline', inplace=True, axis=1)
```

```python
movies.head()
```

### Usunięcie wartości NaN

```python
movies['overview'].isnull().values.any()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(subset=['overview'], inplace=True)
movies.dropna(subset=['poster_path'], inplace=True)
movies.dropna(subset=['original_language'], inplace=True)
movies.dropna(subset=['release_date'], inplace=True)
movies.dropna(subset=['runtime'], inplace=True)
movies.isnull().sum()
```

### Usunięcie odpowiadających wierszy z reszty tabel

```python
# tutaj usuwane sa wiersze z innych kolumn z id filmow, ktorych juz nie ma w movies jkbc
valid_ids = movies['id']

keywords['movieId'] = keywords[keywords['movieId'].isin(valid_ids)]
credits['movieId'] = credits[credits['movieId'].isin(valid_ids)]
# ratings['movieId'] = ratings[ratings['movieId'].isin(valid_ids)]
links['tmdbId'] = links[links['tmdbId'].isin(valid_ids)]
```
